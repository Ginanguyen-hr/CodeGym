```python
class Student:
    def __init__(self, student_id, full_name, dob, hometown, major, class_name):
        self.student_id = student_id
        self.full_name = full_name
        self.dob = dob
        self.hometown = hometown
        self.major = major
        self.class_name = class_name

    def __str__(self):
        return (f"ID: {self.student_id}, Name: {self.full_name}, DOB: {self.dob}, "
                f"Hometown: {self.hometown}, Major: {self.major}, Class: {self.class_name}")

class Student_List:
    def __init__(self):
        self.students = []

    def view_students(self):
        if not self.students:
            print("Danh sách sinh viên hiện đang trống.")
            return
        for stu in self.students:
            print(stu)

    def add_student(self, student):
        # Kiểm tra trùng mã sinh viên
        if any(s.student_id == student.student_id for s in self.students):
            print(f"Không thể thêm. Sinh viên với ID {student.student_id} đã tồn tại.")
            return
        self.students.append(student)
        print(f"Đã thêm sinh viên {student.full_name} vào danh sách.")

    def update_student(self, student_id, **kwargs):

        for stu in self.students:
            if stu.student_id == student_id:
              
                for key, value in kwargs.items():
                    if hasattr(stu, key):
                        setattr(stu, key, value)
                print(f"Đã cập nhật thông tin sinh viên ID {student_id}.")
                return
        print(f"Không tìm thấy sinh viên với ID {student_id} để cập nhật.")

    def delete_student(self, student_id):
        for i, stu in enumerate(self.students):
            if stu.student_id == student_id:
                del self.students[i]
                print(f"Đã xóa sinh viên với ID {student_id}.")
                return
        print(f"Không tìm thấy sinh viên với ID {student_id} để xóa.")

    def search_students(self, keyword):
        keyword_lower = keyword.lower()
        found_students = [stu for stu in self.students if
                          keyword_lower in stu.student_id.lower() or
                          keyword_lower in stu.full_name.lower() or
                          keyword_lower in stu.major.lower() or
                          keyword_lower in stu.class_name.lower()]
        if not found_students:
            print(f"Không tìm thấy sinh viên với từ khóa '{keyword}'.")
            return
        print(f"Tìm thấy {len(found_students)} sinh viên với từ khóa '{keyword}':")
        for stu in found_students:
            print(stu)

    def sort_students(self, by="full_name"):
        if by == "full_name":
            self.students.sort(key=lambda x: x.full_name.lower())
            print("Đã sắp xếp sinh viên theo họ tên.")
        elif by == "student_id":
            self.students.sort(key=lambda x: x.student_id.lower())
            print("Đã sắp xếp sinh viên theo mã sinh viên.")
        else:
            print(f"Không hỗ trợ sắp xếp theo '{by}'.")
```
