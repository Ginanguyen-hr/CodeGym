```python
import turtle
import time
import math

class Clock:
    def __init__(self, hour, minute, second):
        self.hour = hour
        self.minute = minute
        self.second = second
        self.screen = turtle.Screen()
        self.screen.title("Đồng hồ Turtle")
        self.screen.setup(width=600, height=600)
        self.screen.tracer(0)  # Tắt tự động cập nhật màn hình

        self.pen = turtle.Turtle()
        self.pen.hideturtle()
        self.pen.speed(0)

    def draw_clock_face(self):
        self.pen.up()
        self.pen.goto(0, -250)
        self.pen.down()
        self.pen.pensize(5)
        self.pen.circle(250)  
  
        self.pen.pensize(2)
        for i in range(60):
            self.pen.up()
            self.pen.goto(0, 0)
            self.pen.setheading(90)
            self.pen.right(i * 6)
            self.pen.forward(230)
            self.pen.down()
            if i % 5 == 0:
                self.pen.forward(20)  
            else:
                self.pen.forward(10)  
        self.pen.up()

    def draw_hand(self, length, angle, pensize, color):
        self.pen.up()
        self.pen.goto(0, 0)
        self.pen.setheading(90)  
        self.pen.right(angle)
        self.pen.pensize(pensize)
        self.pen.color(color)
        self.pen.down()
        self.pen.forward(length)
        self.pen.up()
        self.pen.goto(0, 0)


    def update(self, hour, minute, second):
        self.hour = hour % 12
        self.minute = minute
        self.second = second

        self.pen.clear()
        self.draw_clock_face()


        angle_sec = second * 6  
        angle_min = minute * 6 + second * 0.1  
        angle_hour = (hour % 12) * 30 + minute * 0.5  

   
        self.draw_hand(100, angle_hour, 6, "black") 
        self.draw_hand(150, angle_min, 4, "blue")    
        self.draw_hand(180, angle_sec, 2, "red")     
        self.screen.update()
clock = Clock(0, 0, 0)
clock.draw_clock_face()

while True:
    now = time.localtime()  
    clock.update(now.tm_hour, now.tm_min, now.tm_sec)
    time.sleep(1)
```
