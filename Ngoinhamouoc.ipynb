```python
import turtle
def draw_rectangle(t, width, height, color):
    t.begin_fill()
    t.fillcolor(color)
    for _ in range(2):
        t.forward(width)
        t.left(90)
        t.forward(height)
        t.left(90)
    t.end_fill()
def draw_triangle(t, base, height, color):
    t.begin_fill()
    t.fillcolor(color)
    t.forward(base)
    t.left(135)
    t.forward((2**0.5)*height)
    t.left(90)
    t.forward((2**0.5)*height)
    t.left(135)
    t.end_fill()
def move_to(t, x, y):
    t.penup()
    t.goto(x, y)
    t.pendown()
def draw_house(t, x, y):
    move_to(t, x, y)
    draw_rectangle(t, 200, 150, "burlywood")
    move_to(t, x, y + 150)
    t.begin_fill()
    t.fillcolor("brown")
    t.goto(x + 100, y + 250)
    t.goto(x + 200, y + 150)
    t.goto(x, y + 150)
    t.end_fill()
    move_to(t, x + 70, y)
    draw_rectangle(t, 60, 90, "saddlebrown")
    move_to(t, x + 20, y + 80)
    draw_rectangle(t, 40, 40, "skyblue")
    move_to(t, x + 140, y + 80)
    draw_rectangle(t, 40, 40, "skyblue")
def draw_tree(t, x, y):
    move_to(t, x, y)
    draw_rectangle(t, 20, 60, "saddlebrown")
    move_to(t, x - 20, y + 60)
    t.begin_fill()
    t.fillcolor("forestgreen")
    t.circle(30)
    t.end_fill()
def main():
    screen = turtle.Screen()
    screen.bgcolor("lightblue")
    pen = turtle.Turtle()
    pen.speed(0)
    move_to(pen, -400, -150)
    draw_rectangle(pen, 800, 200, "green")
    draw_house(pen, -100, -150)
    draw_tree(pen, 150, -150)
    draw_tree(pen, -200, -150)
    draw_person(pen, 0, -60)
    pen.hideturtle()
    screen.mainloop()
main()
```
