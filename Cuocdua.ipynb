```python
import turtle
import random
import time
from datetime import datetime

screen = turtle.Screen()
screen.title("üê¢ Cu·ªôc ƒëua R√πa üèÅ")
screen.bgcolor("lightgreen")
screen.setup(width=800, height=400)


finish_line = 300
line = turtle.Turtle()
line.penup()
line.goto(finish_line, 150)
line.pendown()
line.goto(finish_line, -150)
line.hideturtle()


racer1 = turtle.Turtle()
racer1.color("red")
racer1.shape("turtle")
racer1.penup()
racer1.goto(-300, 50)
racer1.pendown()

racer2 = turtle.Turtle()
racer2.color("blue")
racer2.shape("turtle")
racer2.penup()
racer2.goto(-300, -50)
racer2.pendown()


message = turtle.Turtle()
message.penup()
message.hideturtle()
message.goto(0, 100)
message.color("black")

start_time = None
race_running = False

def start_race():
    global start_time, race_running
    if race_running:
        return
    race_running = True
    start_time = datetime.now()

    while True:
   
        racer1.forward(random.randint(1, 10))
        racer2.forward(random.randint(1, 10))
        time.sleep(0.05)

       
        if racer1.xcor() >= finish_line:
            end_time = datetime.now()
            duration = (end_time - start_time).total_seconds()
            message.write("üèÜ R√πa ƒê·ªé th·∫Øng cu·ªôc!\n‚è±Ô∏è Th·ªùi gian: {:.2f} gi√¢y".format(duration), align="center", font=("Arial", 16, "bold"))
            break
        elif racer2.xcor() >= finish_line:
            end_time = datetime.now()
            duration = (end_time - start_time).total_seconds()
            message.write("üèÜ R√πa XANH th·∫Øng cu·ªôc!\n‚è±Ô∏è Th·ªùi gian: {:.2f} gi√¢y".format(duration), align="center", font=("Arial", 16, "bold"))
            break


screen.listen()
screen.onkey(start_race, "space")

hint = turtle.Turtle()
hint.penup()
hint.hideturtle()
hint.goto(0, 160)
hint.write("Nh·∫•n ph√≠m [Space] ƒë·ªÉ b·∫Øt ƒë·∫ßu cu·ªôc ƒëua!", align="center", font=("Arial", 14, "normal"))

screen.mainloop()
```
